```python
#Load dataset 
import pandas as pd

df = pd.read_csv("../data/processed/combined_spam_dataset.csv")

print(df.shape)
df.head()
```

```python
#Basic Cleaning
import re

def clean_text(text):
    text = str(text)
    
    # remove extra spaces
    text = text.strip()
    
    # remove multiple spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text

df["text"] = df["text"].apply(clean_text)
```

```python
#Handling missing values
df = df.dropna(subset=["text", "label"])
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Text Length Feature (Optional but Useful)
df["message_length"] = df["text"].apply(len)
```

```python
# Final verification
print(df.shape)
print(df.isnull().sum())
print(df["label"].value_counts())
```

```python
print(df["source"].value_counts())
```

```python
#Save cleaned dataset
df.to_csv("../data/processed/cleaned_spam_dataset.csv", index=False)
```
